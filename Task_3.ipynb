```python
sales_update=sales.drop(['Country','Postal Code'],axis=1)
sales_update.head(10)
```

# Region vise Analysis

```python
sales_update.groupby(by='Region')[['Profit','Sales']].sum().plot.bar(color=['orange','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

### West region has highest profit and sales
### Central region has third highest sales but least profit.

# State vise Analysis

```python
ps = sales_update.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps.plot.bar(color=['Red','blue'],figsize=(25,15))
plt.title('Profit/Loss and Sales across states')
plt.xlabel('States')
plt.ylabel('profit/Loss and Sales')
plt.show()
```

### California has highest profit and sales.
### Whereas,Texas has third highest sales but lowest profit.

# City vise Analysis

```python
g = sales.groupby('City')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False)
g.plot.bar(color=['orange','maroon'],figsize=(25,18))
```

```python
g.max
```

```python
g.min
```

### New york City has Highest profit and sales.
### Philadelphia has least profit.
### Abilene has lowest sales.

# Ship-Mode Counts

```python
sns.countplot(sales_update['Ship Mode'])
plt.show()
```

### Standard class ship-mode is highest use by the customers.

# Profit with respect to category

```python
plt.title("%Profit with respect to Category")
plt.pie(sales_update['Profit'].groupby(sales_update['Category']).sum(),labels=sales_update['Profit'].groupby(sales_update['Category']).sum().index,autopct='%0.1f%%')
plt.show()
```

### Technology category has highest profit on the other hand Furniture has lowest.

# Profit with respect to Sub-Category

```python
plt.figure(figsize=(15,7))
plt.title("Profit vs Sub-Category ")
sns.barplot(x='Sub-Category',y='Profit',data=sales_update)
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.countplot(sales_update['Sub-Category'])
plt.show()
```

### Copiers sub-category has highest profit whereas tables has lowest.
### Binders sub-category has highest count whereas Copiers has lowest.

# Profit with respect to Discount

```python
sns.lineplot(data=sales_update, x="Discount", y="Profit")
plt.show()
```

### If Seller give 0.1% discount then they will get highest profit however, if seller give more than 0.1% discount then they will be in loss.

# profit with respect to segment

```python
plt.title("%Profit with segment")
plt.pie(sales_update['Profit'].groupby(sales_update['Segment']).sum(),labels=sales_update['Profit'].groupby(sales_update['Segment']).sum().index,autopct='%0.1f%%')
plt.show()
```

### Consumer segment has more profit comapare to other segments.
