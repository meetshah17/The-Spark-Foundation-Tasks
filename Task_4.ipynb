## Taking whole dataset partialy and Cleaned it
### Taking around 10-15 columns individually analyse whole dataset and remove unnecessary columns.
### Here we use iloc function to get information partialy and if there is any null value or unnecessary columns are present then we'll drop it.

```python
sd = tsf.iloc[:,:10]
sd
```

```python
final_ed = sd.drop(['approxdate','resolution','extended'],axis=1)
final_ed
```

```python
fd = tsf.iloc[:,10:26]
fd
```

```python
final_jd = fd.drop(['multiple'],axis=1)
final_jd
```

```python
gd = tsf.iloc[:,26:41]
gd
```

```python
final_hd = gd.drop(['suicide','attacktype2','attacktype2_txt','attacktype3','attacktype3_txt'],axis=1)
final_hd
```

```python
ad = tsf.iloc[:,41:56]
ad
```

```python
final_td = ad.drop(tsf.iloc[:,42:56],axis = 'columns')
final_td
```

```python
ud = tsf.iloc[:,56:71]
ud
```

```python
bg = ud.drop(['natlty3','natlty3_txt'],axis=1)
bg
```

```python
tg = bg.drop(tsf.iloc[:,59:69],axis=1)
tg
```

```python
final_tf = tg.drop(['nperpcap'],axis=1)
final_tf
```

```python
md = tsf.iloc[:,71:86]
md
```

```python
td = md.drop(tsf.iloc[:,73:81],axis=1)
td
```

```python
final_rd = td.drop(['weaptype2'],axis=1)
final_rd
```

```python
re = tsf.iloc[:,86:101]
re
```

```python
ew = re.drop(tsf.iloc[:,86:97],axis=1)
ew
```

```python
final_me = ew.drop(['nkillus','nkillter'],axis=1)
final_me
```

```python
ye = tsf.iloc[:,101:116]
ye
```

```python
final_op = ye.drop(['nwoundte','nwoundus','propextent_txt','nhours','ndays','nhostkid','nhostkidus','divert','ishostkid'],axis=1)
final_op
```

```python
pe = tsf.iloc[:,116:]
pe
```

```python
rq = pe.drop(tsf.iloc[:,118:126],axis=1)
rq
```

```python
final_qw = rq.drop(['related'],axis=1)
final_qw
```

## Clean Dataset
### Concatenate whole clean dataset together and merge it in a one variable.

```python
final_con = pd.concat([final_ed,final_jd,final_hd,final_td,final_tf,final_rd,final_me,final_op,final_qw],axis=1)
final_con # clean dataset.
```

```python
final_con.columns
```

# Data Visualization and Analysis

```python
# terrorist attack by region from 1970 to 2015
plt.figure(figsize=(20,10))
final_con.groupby(by='iyear')['region'].sum().sort_values(ascending=True).plot(kind='bar')
plt.show()
```

### we can see that in 2014 highest attack occurs.

```python
plt.figure(figsize=(20,10))
final_con.groupby(by='gname')['success'].sum().sort_values(ascending=True).plot(kind='line')
```

### hezbollah gang success ratio is maximum then anyother gang.

```python
plt.figure(figsize=(20,10))
final_con.groupby(by='country_txt')['region'].sum().sort_values(ascending=True).plot(kind='bar')
```

```python
m = final_con.groupby(by='country_txt')['region'].sum().sort_values(ascending=True)
m.max
```

### here we can see that Iraq has a maximum attacks.
## Take whole dataset of Iraq and analyse it.

```python
iraq = final_con.loc[(final_con['country_txt'] =='Iraq')]
iraq
```

```python
plt.figure(figsize=(20,10))
final_con.groupby(by='region_txt')['region'].sum().sort_values(ascending=True).plot(kind='bar',color='#f10146')
```
